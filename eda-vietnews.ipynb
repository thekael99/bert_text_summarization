```python
import pandas as pd
import numpy as np

from glob import glob
from functools import reduce
import re
from tqdm import tqdm
```

```python
def format_file(path):
    f = open(path, mode="r", encoding="utf8")
    text = [x.replace("\n", " ") for x in f.readlines()]
    f.close()
    res = {}
    res["Title"] = text[0].replace("_", " ")
    res["Abstract"] = text[2].replace("_", " ")
    content = []
    for sentence in text[4:]:
        if sentence == "":
            break
        content.append(sentence.replace("_", " "))
#         content = content + " " + sentence.replace("_", " ")
    res["Content"] = content
    return res
```

```python
file_path = sorted(glob("../input/vietnews/vietnews/data/train_tokenized/*"))
print(len(file_path))
```

```python
df = pd.DataFrame([format_file(x) for x in tqdm(file_path[:100])])
df
```

```python
df = df.drop(["Abstract"], axis=1)
df.rename(columns={"Title": "tgt", "Content": "src"}, inplace=True)

df.dropna(inplace=True)
df = df[["src", "tgt"]]

df["src"] = df["src"].apply(lambda sents: [sent.split(" ")[:-1] for sent in sents])
df["tgt"] = df["tgt"].apply(lambda sent: [sent.split(" ")[:-1]])
df
```
